```python
from selenium import webdriver
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(browser.page_source,'html.parser')
```

```python
sp=soup.find('ytd-rich-grid-renderer')
```

```python
sp
```

```python
sp.find_all('ytd-rich-item-renderer')[1].text
```

```python
sp1=sp.find_all('ytd-rich-item-renderer')
```

```python
sp1[0]
```

```python
data=[]
for i in sp1:
    title=i.find('a',class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').text
    link=i.find('a',class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').get('href')
    views=i.find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')[0].text
    date=i.find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')[1].text
    thumbnail=sp.find('img').get('src').split('?')[0]
    
    data.append([title,link,views,date,thumbnail])
```

```python
len(data)
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=['title','link','views','date','thumbnail'])
```

```python
df.to_csv('youtube1.csv')
```

```python
import pandas as pd
df=pd.read_csv("youtube1.csv")
import time
from bs4 import BeautifulSoup
from tqdm import tqdm
import numpy as np
```

```python
data=[]
for i in tqdm(df["link"]):
    link='https://www.youtube.com'+i
    browser.get(link)
    time.sleep(12)
    soup=BeautifulSoup(browser.page_source,'html.parser')
    try:
        views=soup.find_all('span',dir='auto')[0].text
    except:
        views=np.nan
    try:
        date=soup.find_all('span',dir='auto')[2].text
    except:
        date=np.nan
    try:
        title=soup.find('yt-formatted-string',class_='style-scope ytd-watch-metadata').text
    except:
        title=np.nan
    try:
        likes=soup.find('span',class_="yt-core-attributed-string yt-core-attributed-string--white-space-pre-wrap").text
    except:
        likes=np.nan
    try:
        description=soup.find('span',class_='yt-core-attributed-string--link-inherit-color').text
    except:
        description=np.nan
    data.append([title,date,likes,views,description])
    
    
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=['title','data','likes','views','description'])
```
