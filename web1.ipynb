```python
data=[]
for s in soup.find_all('div',class_='quote'):
    quote=s.find('span',class_='text').text[1:-1]
    author=s.find('small').text
    author_id=s.find('a').get('href')
    tags=[]
    for i in s.find_all('a',class_='tag'):
        tags.append(i.text)
    tags=','.join(tags)
    
    print(quote)
    print(author)
    print(author_id)
    print(tags)
    print("-"*40)
    data.append([quote,author,author_id,tags])
```

```python
from tqdm import tqdm
```

```python
data=[]
for page in tqdm(range(1,11)):
    link=('https://quotes.toscrape.com/page/'+str(page))
    res=requests.get(link)
    soup=BeautifulSoup(res.text,'html.parser')
    for s in soup.find_all('div',class_='quote'):
        quote=s.find('span',class_='text').text[1:-1]
        author=s.find('small').text
        author_id=s.find('a').get('href')
        tags=[]
        for i in s.find_all('a',class_='tag'):
            tags.append(i.text)
    tags=','.join(tags)
    
    
    data.append([quote,author,author_id,tags])

    
```

```python
df=pd.DataFrame(data,columns=['quote','author','author_id','tags'])
```
